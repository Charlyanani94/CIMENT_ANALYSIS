```python
# Sélection unique des colonnes numériques
numeric_data = data.select_dtypes(include=[np.number])

# Calcul  de la matrice de corrélation
correlation_matrix = numeric_data.corr()

# Filtrage unique des corrélations fortes positives (≥ 0.7)
strong_corr_pairs = []

for col in correlation_matrix.columns:
    for idx in range(len(correlation_matrix)):
        if col != correlation_matrix.index[idx]:  # Éviter la corrélation avec soi-même
            corr_value = correlation_matrix[col].iloc[idx]  # Utiliser .iloc pour accéder par position
            if corr_value >= 0.7:  # Garder uniquement les corrélations positives
                strong_corr_pairs.append((col, correlation_matrix.index[idx], corr_value))

# Affichage des paires de variables avec leur corrélation
if not strong_corr_pairs:
    print("Aucune corrélation forte positive détectée (≥ 0.7).")
else:
    for var1, var2, corr in strong_corr_pairs:
        print(f"{var1} & {var2} : {corr:.2f}")
```

```python
# Fonction pour choisir entre la moyenne et la médiane
def impute_with_mean_or_median(series):
    if series.isnull().sum() == 0:
        return series 
    # Calcule de la moyenne et la médiane
    mean_value = series.mean()
    median_value = series.median()
    
    # Vérification de la symétrie de la distribution
    skewness = series.skew()
    
    # Choix de la méthode d'imputation
    if abs(skewness) < 0.5:  # Distribution symétrique
        return series.fillna(mean_value)
    else:  # Distribution asymétrique
        return series.fillna(median_value)

# Application de la fonction d'imputation aux colonnes 
data['calcium_libre'] = impute_with_mean_or_median(data['calcium_libre'])
```

```python
from sklearn.preprocessing import StandardScaler

# Détection des valeurs aberrantes (exemple avec IQR)
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

# Traitement des valeurs aberrantes dans le DataFrame 'data'
for column in data.select_dtypes(include=[np.number]).columns:  # Filtrer uniquement les colonnes numériques
    outliers = detect_outliers_iqr(data, column)
    if not outliers.empty:
        median_value = data[column].median()
        # Remplacer les valeurs aberrantes par la médiane
        data[column] = np.where((data[column] < outliers[column].min()) | (data[column] > outliers[column].max()), median_value, data[column])

# Standardisation
scaler = StandardScaler()
standardized_data = scaler.fit_transform(data.select_dtypes(include=[np.number]))  # Standardiser uniquement les colonnes numériques

# Convertir en DataFrame
standardized_df = pd.DataFrame(standardized_data, columns=data.select_dtypes(include=[np.number]).columns)

# Afficher le DataFrame standardisé
print(standardized_df)
```
